```python
# Importar as Bibliotecas ##
import pytesseract #estamos usando a versão 0.3.9. Caso queira verificar qual versão está usando, passe o seguinte código 'pytesseract.__version__'
import cv2 #estamos usando a versão 0.3.9. Caso queira verificar qual versão está usando, passe o seguinte código 'cv2.__version__'

import os
import matplotlib.pyplot as plt
```

```python
print(os.getcwd())
```

```python
path = 'Images/4.jpeg'
img = cv2.imread(path)
```

```python
plt.imshow(img);
plt.axis("off");
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap="gray");
plt.axis("off");
```

```python
config_tesseract = '--psm 1' # Assume a single column of text of variable sizes.
texto = pytesseract.image_to_string(gray, lang='por', config=config_tesseract)

print(texto)
```
